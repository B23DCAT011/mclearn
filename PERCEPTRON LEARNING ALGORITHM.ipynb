```python
import numpy as np
def predict(w,X):
    return np.sign(X.dot(w))

def perceptron(X,y,w_init):
    w = w_init
    while True:
        pred = predict(w,X)
        mis_idxs = np.where(np.equal(pred, y) == False)[0]
        num_mis = mis_idxs.shape[0]
        if num_mis == 0:
            return w
        random_id = np.random.choice(mis_idxs,1)[0]
        w = w + y[random_id]*X[random_id]
```

```python
 means=[[-1,0],[1,0]]
 cov=[[.3,.2],[.2,.3]]
 N=10
#tao 2 mien du lieu rieng biet
 X0=np.random.multivariate_normal(means[0],cov,N)
 X1=np.random.multivariate_normal(means[1],cov,N)
#ghep 2 cái vào 1 ma trận,tạo ra linearly separable
 X=np.concatenate((X0,X1),axis=0)
print(X.shape)
#LABEL
 y=np.concatenate((np.ones(N),-1*np.ones(N)))
print(y)
 Xbar=np.concatenate((np.ones((2*N,1)),X),axis=1)#add bias
print(Xbar.shape)
 w_init=np.random.randn(Xbar.shape[1])
 w=perceptron(Xbar,y,w_init)
```

```python
import matplotlib.pyplot as plt

# Vẽ các điểm dữ liệu
plt.scatter(X[:N, 0], X[:N, 1], color='blue', label='Class 1')
plt.scatter(X[N:, 0], X[N:, 1], color='red', label='Class -1')

# Vẽ siêu phẳng phân chia
x_vals = np.linspace(-2, 2, 100)
y_vals = -(w[1] * x_vals + w[0]) / w[2]
plt.plot(x_vals, y_vals, label='Decision boundary')

plt.legend()
plt.show()
```
