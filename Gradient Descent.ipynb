```python
import numpy as np
import matplotlib.pyplot as plt

def gradient_descent(x, y, learning_rate=0.01, iterations=10000):
    m_cur = b_cur = 0
    n = len(x)  # Số lượng mẫu dữ liệu

    for i in range(iterations):
        y_predict = m_cur * x + b_cur  # Tính giá trị dự đoán
        cost = np.mean((y - y_predict) ** 2)  # Hàm cost (MSE)

        # Tính đạo hàm
        md = -(2/n) * np.dot(x, (y - y_predict))
        bd = -(2/n) * np.sum(y - y_predict)

        # Cập nhật giá trị
        m_cur -= learning_rate * md
        b_cur -= learning_rate * bd

        # In kết quả mỗi 100 vòng
        if i % 100 == 0:
            print(f"Iteration {i}: m={m_cur:.5f}, b={b_cur:.5f}, cost={cost:.5f}")

    return m_cur, b_cur

# Dữ liệu đầu vào (CHUẨN HÓA)
x = np.array([92, 56, 88, 70, 80, 49, 65, 35, 66, 67], dtype=np.float64)
y = np.array([98, 68, 81, 80, 83, 52, 66, 30, 68, 73], dtype=np.float64)

x = x/100  # CHUẨN HÓA DỮ LIỆU
y = y/100 

# Chạy thuật toán
m_final, b_final = gradient_descent(x, y)

print(f"\nFinal result: m={m_final:.5f}, b={b_final:.5f}")

# Vẽ biểu đồ
plt.scatter(x, y, color='blue', label="Original data")
plt.plot(x, m_final*x + b_final, color='red', label="Fitted line")
plt.xlabel("X values (normalized)")
plt.ylabel("Y values (normalized)")
plt.legend()
plt.show()
```
